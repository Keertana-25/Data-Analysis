#### Data Analysis
The average final grade was calculated using the G3 column. The number of students scoring above 15 was identified. Correlation analysis was used to study the relationship between study time and performance. Gender-wise average grades were also calculated.

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['G3'])
plt.xlabel("Final Grade(G3")
plt.ylabel("Number Of Students")
plt.title("Distribution Of Final Grades")
plt.show()
```

## Histogram of Final Grades
A histogram was plotted to understand the distribution of final grades among students.


```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

## Study Time vs Final Grade
A scatter plot was created to visualize the relationship between study time and final grades.

```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Average Grade by Gender")
plt.show()
```

## Average Grade by Gender
A bar chart was plotted to compare the average final grade of male and female students.
